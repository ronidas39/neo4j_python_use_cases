# tutorial 90,how to load json data from multiple url using cypher,apoc and python

```python
from neo4j import GraphDatabase
import io
driver=GraphDatabase.driver(uri="bolt://localhost:7687",auth=("neo4j","Rambo@1234"))
session=driver.session()
```

```python
with io.open("urls.txt","r",encoding="utf-8")as f1:
    file_data=f1.read()
    f1.close()
file_data=file_data.split("\n")
file_data
```

```python
q1="""
call apoc.load.josn($url)
yield value
"""
```

```python
json_data=[]
for url in file_data:
    x={"url":url}
    results=session.run(q1,x)
    data=results.data()
    json_data.append(data)
    
```

```python
for row in json_data:
    print(row)
    print("__________")
```
